# Spatial mapping of cell types to NanostringWTA data of 19pcw and 14pcw fetal brains with cell2location
In this notebook we we map fetal brain cell types to regions of interest (ROIs) profiled with the NanostringWTA technology, using a version of our cell2location method recommended for probe based spatial transcriptomics data.

Initialize path to this github repository:

```python
import sys,os
path = '/nfs/team283/aa16/KR_NAS/'
os.chdir(path)
sys.path.append(path)
```

Load the required modules and configure theano settings:

```python
import pickle
import anndata
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data_type = 'float32'
os.environ["THEANO_FLAGS"] = 'device=cuda,floatX=' + data_type + ',force_device=True' + ',dnn.enabled=False'
import isctools
```

Load data from our NanostringWTA run:

```python
adata_wta = pickle.load(open(path + "data/nanostringWTA_fetailBrain_AnnData.p", "rb" ))
```

We have already annotated ROIs as good or low quality based on a read per nuclei count above 2000 and we do not use low quality ROIs in the analysis here:

```python
adata_wta = adata_wta[adata_wta.obs['GoodQuality'],:]
```

Here we load an existing matrix of average gene expression profiles for each cell type expected in our nanostringWTA data (taken from the single-cell RNAseq study Polioudakis et al., Neuron, 2019):

```python
meanExpression_sc = pd.read_csv(path + "data/polioudakis2019_meanExpressionProfiles.csv", index_col=0)
```

Make sure our nanostringWTA data and single-cell RNAseq reference have the same genes (excluding negative probes):

```python
wta_genes = np.asarray(adata_wta.var_names)
common_genes = [wta_genes[i] for i in range(len(wta_genes)) if wta_genes[i] in meanExpression_sc.index] 
adata_wta = adata_wta[:,[adata_wta.var_names[i] in common_genes + ['NegProbe-WTX'] for i in range(len(adata_wta.var_names))]]
meanExpression_sc = meanExpression_sc.reindex(index = common_genes)
```

The model needs seperate gene probes and negative probes in two numpy arrays and if available we can also supply nuclei counts. The average gene expression formats should also be a numpy array. We initialize all of those here:

```python
counts_geneProbes = np.asarray(adata_wta[:,np.array(adata_wta.var != 'NegProbe-WTX').squeeze()].X)
counts_negativeProbes = np.asarray(adata_wta[:,np.array(adata_wta.var == 'NegProbe-WTX').squeeze()].X)
counts_nuclei = np.asarray(adata_wta.obs['nuclei']).reshape(len(adata_wta.obs['nuclei']),1)
celltypes = meanExpression_sc.keys()
meanExpression_sc = np.asarray(meanExpression_sc)
```

This histogram shows the range of nuclei counts we have in our ROIs:

```python
plt.hist(counts_nuclei, bins = 20)
plt.xlabel('Number of Nuclei')
plt.ylabel('Occurences')
plt.title('Number of Nuclei in ROIs')
```

We took those ROIs at different positions across the temporal cortex (cortical depth), with a focus on the ventricular and subventricular zone at low cortical depth, where we also collected cell type specific areas of interest:

```python
plt.hist(adata_wta.obs['VCDepth'], bins = 5)
plt.xlabel('Cortical Depth')
plt.ylabel('Occurences')
plt.title('Number of ROIs at different Cortical Depth')
```

The majority of our ROIs comes from two 19pcw cases, but we also have a 14pcw case:

```python
plt.hist(adata_wta.obs['age'], bins = 3)
plt.xlabel('Cortical Depth')
plt.ylabel('Occurences')
plt.title('Number of ROIs per Gestational Age')
```

Plotting the number of reads vs the number of nuclei in each ROIs and colouring the dots by the order of ROI collection (dark blue at the beginning, bright yellow at the end), shows some batch effects in sensitivity that depend on collection time. These are currently not accounted for by our model. In general, if these effects are large it could be better to fit seperate models to seperate batches, but we don't do this here.

```python
plt.show()
plt.scatter(adata_wta.obs['nuclei'], adata_wta.obs['AlignedReads'],
           c = adata_wta.obs['ORDER'])
plt.xlabel('Number of Nuclei')
plt.ylabel('Number of Reads')
plt.title('Reads/Nuclei coloured by collection order')
```

Initialize the model:

```python
mod1 = isctools.ProbeCounts_CoLocationModel_Poisson(
        cell_state_mat = meanExpression_sc,
        X_data = counts_geneProbes,
        Y_data = counts_negativeProbes,
        data_type='float32',
        n_iter=20000,
        learning_rate=0.001,
        fact_names=celltypes,
        gene_level_prior={'mean': 1/2, 'sd': 1/8, 'mean_var_ratio': 0.25},
        cell_number_prior={'cells_per_spot': counts_nuclei, 'factors_per_spot': 6, 'combs_per_spot': 50},
        cell_number_var_prior={'cells_mean_var_ratio': 0.1, 'factors_mean_var_ratio': 1, 'combs_mean_var_ratio': 1})
```

Fit the model with two different initializations (n = 2):

```python
mod1.fit_advi_iterative(n_iter = 30000, learning_rate = 0.001, n=2, method='advi')
```

```python
mod1.fit_advi_refine(n_iter=5000, learning_rate = 0.000001)
```

The above plots the optimization loss from the last 5000 iterations, which clearly shows that it has converged. In addition, here is the full optimzation history:

```python
mod1.plot_history()
```

We can check the stability of our optimization by comparing the two different initalizations. We can do this by looking at the correlation of cell type density across locations between the two initializations. Given each cell type seems to be highly correlated with itself (yellow diagonal), we can conclude that the optimzation produced similar results in both cases.

```python
mod1.evaluate_stability()
```

Sample posterior and plot posterior mu vs data:

```python
mod1.sample_posterior(node='all', n_samples=1000, save_samples=False);
```

```python
mod1.sample2df(node_name='spot_factors')
```

```python
isctools.ProbeCounts_CoLocationModel_Poisson.compute_expected(mod1)
mod1.plot_posterior_mu_vs_data()
```

We have some function ready for plotting cell locations in 1D. We can use them here to plot the 19pcw and 14pcw cell locations for all 16 cell types. This one is a scatter plot, with a polynomial best fit line, that optionally plots either the total number or density of cells:

```python
x = np.asarray(adata_wta.obs['VCDepth'])
order_1 = np.where([adata_wta.obs['slide'][i] in ('00MV', '00MV-2', '00MU') and adata_wta.obs['Radial_position'][i] == 2 for i in range(len(adata_wta.obs['slide']))])[0]
order_1 = order_1[np.argsort(x[order_1])]

x = np.asarray(adata_wta.obs['VCDepth'])
order_2 = np.where([adata_wta.obs['slide'][i] in ('00MW') and adata_wta.obs['Radial_position'][i] == 2 for i in range(len(adata_wta.obs['slide']))])[0]
order_2 = order_2[np.argsort(x[order_2])]

order = [order_1, order_2]
```

```python
isctools.plot_Locations_1D_scatterPlot(mod1, x, order, 
                              xlabel = 'Cortical Depth',
                              categories = ('19pcw', '14pcw'),
                              density = False,
                              polynomial_order=6)
```

```python
isctools.plot_Locations_1D_scatterPlot(mod1, x, order, 
                              xlabel = 'Cortical Depth',
                              categories = ('19pcw', '14pcw'),
                              density = True,
                              polynomial_order=6)
```

Importantly, vRG and oRG localize in the ventricular and subventricual zone respectively and thus exactly agree with prior knowledge. Checking the model reproduces such known distinct cell type localizations is generally recommended, before looking at new biological insights.
