```python
def evaluatePerformance(sc_results, wta_results):
    
    n_genes = sum(wta_results['FDR'] < 0.05)
    n_genes_oRG = sum([wta_results['mean'][i] < 0 and wta_results['FDR'][i] < 0.05 for i in range(len(wta_results['FDR']))])
    n_genes_IP = sum([wta_results['mean'][i] > 0 and wta_results['FDR'][i] < 0.05 for i in range(len(wta_results['FDR']))])
    
    sc_genes = np.array(sc_results['gene'])
    sc_results = sc_results.iloc[[sc_results['gene'].iloc[i] in wta_results.index for i in range(len(sc_results['gene']))],:]
    wta_results = wta_results.iloc[[wta_results.index[i] in sc_genes for i in range(len(wta_results.index))],:]
    wta_IP_genes = wta_results.iloc[[wta_results['mean'].iloc[i] > 0 and wta_results['FDR'].iloc[i] < 0.05 for i in range(len(wta_results['FDR']))],:].index
    wta_oRG_genes = wta_results.iloc[[wta_results['mean'].iloc[i] < 0 and wta_results['FDR'].iloc[i] < 0.05 for i in range(len(wta_results['FDR']))],:].index
    sc_oRG_genes = np.array(sc_results.iloc[[sc_results['log2fc'].iloc[i] > 1 and sc_results['qval'].iloc[i] < 0.05 for i in range(len(sc_results['qval']))],:]['gene'])
    sc_IP_genes = np.array(sc_results.iloc[[sc_results['log2fc'].iloc[i] < 1 and sc_results['qval'].iloc[i] < 0.05 for i in range(len(sc_results['qval']))],:]['gene'])
    
    if len(wta_IP_genes) + len(wta_oRG_genes) != 0:
        recall = (sum([wta_oRG_genes[i] in sc_oRG_genes for i in range(len(wta_oRG_genes))]) +
    sum([wta_IP_genes[i] in sc_IP_genes for i in range(len(wta_IP_genes))]))/(len(sc_oRG_genes) + len(sc_IP_genes))
        precision = (sum([wta_oRG_genes[i] in sc_oRG_genes for i in range(len(wta_oRG_genes))])+
                 sum([wta_IP_genes[i] in sc_IP_genes for i in range(len(wta_IP_genes))]))/(len(wta_IP_genes) + len(wta_oRG_genes))
        fscore = 2*(precision*recall)/(precision+recall)
    else:
        precision = 0
        recall = 0
        fscore = 0
    
    if len(wta_oRG_genes) != 0:
        recall_oRG = sum([wta_oRG_genes[i] in sc_oRG_genes for i in range(len(wta_oRG_genes))])/len(sc_oRG_genes)
        precision_oRG = sum([wta_oRG_genes[i] in sc_oRG_genes for i in range(len(wta_oRG_genes))])/len(wta_oRG_genes)
        fscore_oRG = 2*(precision_oRG*recall_oRG)/(precision_oRG+recall_oRG)
    else:
        recall_oRG = 0
        precision_oRG = 0
        fscore_oRG = 0
    
    
    if len(wta_IP_genes) != 0:
        recall_IP = sum([wta_IP_genes[i] in sc_IP_genes for i in range(len(wta_IP_genes))])/len(sc_IP_genes)
        precision_IP = sum([wta_IP_genes[i] in sc_IP_genes for i in range(len(wta_IP_genes))])/len(wta_IP_genes)
        fscore_IP = 2*(precision_IP*recall_IP)/(precision_IP+recall_IP)
    else:
        recall_IP = 0
        precision_IP = 0
        fscore_IP = 0
    
    return np.array(((fscore, precision, recall, n_genes), (fscore_oRG, precision_oRG, recall_oRG, n_genes_oRG), (fscore_IP, precision_IP, recall_IP, n_genes_IP)))
```

```python
sc_results = pickle.load(open('/nfs/team283/aa16/KR_NAS/data/oRG_IP_scRNAseq_DE_cpm_results.p', "rb"))

all_performances = [None] * 12
for x in range(12):
    all_performances[x] = []
    for i in range(12):
        print(i)
        try:
            wta_results = pickle.load(open("/nfs/team283/aa16/KR_NAS/savedModels/benchmarking/WTACounts_GeneralModel_Results_parameterSet" + str(x) + "n_iter" + str(n_iterations[i]) + ".p", 'rb' ))
            all_performances[x].append(evaluatePerformance(sc_results, wta_results))
        except:
            print('skipping')
```
