In this notebook we show which ligands and receptors are detected in EOMES and HOPX AOIs, compared to single-cell RNAseq data from Polioudakis 2019:

```python
import sys,os
path = '/nfs/team283/aa16/KR_NAS/'
os.chdir(path)
sys.path.append(path)
```

```python
%pylab inline
import pandas as pd
import NaiveDE
import scipy.stats as ss
import pickle as pickle
import scipy as sc
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
```

Load the data and remove negative probes for this analysis:

```python
adata = pickle.load(open("/nfs/team283/aa16/KR_NAS/data/nanostringWTA_fetailBrain_AnnData.p",  "rb"))
adata = adata[:,adata.var_names != 'NegProbe-WTX']
# adata.obs['TotalCounts_corrected'] = np.sum(adata.layers['X_Corrected'], axis = 1)
# adata.layers['CPM_corrected'] = np.asarray([adata.layers['X_Corrected'][i,:]/adata.obs['TotalCounts_corrected'][i] for i in range(len(adata.obs['TotalCounts_corrected']))])*10**6
```

Remove too high and too low expressed genes:

```python
plt.scatter(range(np.shape(adata.X)[1]), np.mean(adata.X, axis = 0), label = 'Before')
plt.yscale('log')
plt.xlabel('genes')
plt.ylabel('average counts')

adata = adata[:, np.mean(adata.X, axis = 0) < 3000]
adata = adata[:, np.mean(adata.X, axis = 0) > 10]

plt.scatter(range(np.shape(adata.X)[1]), np.mean(adata.X, axis = 0), label = 'After')
plt.yscale('log')
plt.xlabel('genes')
plt.ylabel('average counts')
plt.legend()
```

Load polioudakis data as a reference and only keep genes that are common between wta and reference:

```python
polioudakis = pd.read_csv("../InSituCellTools/data/polioudakis2019_meanExpressionProfiles.csv", index_col=0)
```

```python
polioudakis
```

```python
wta_genes = np.asarray(adata.var_names)
common_genes = [wta_genes[i] for i in range(len(wta_genes)) if wta_genes[i] in polioudakis.index] 
adata = adata[:,[adata.var_names[i] in common_genes + ['NegProbe-WTX'] for i in range(len(adata.var_names))]]
adata = adata[:,[np.where(np.array(common_genes) == g)[0][0] for g in adata.var_names if g in common_genes]]
polioudakis = polioudakis.reindex(index = common_genes)
```

```python
polioudakis
```

Subset to ligand-receptor pairs:

```python
cellPhoneDB_geneInput = pd.read_csv('../data/CellPhoneDB/gene_input.csv')
subset_genes = np.array(cellPhoneDB_geneInput['gene_name'])
subset = [adata.var_names[i] in subset_genes for i in range(len(adata.var_names))]
```

Find detected genes in HOPXpos and EOMESpos AOIs (detected in at least 2 samples), as well as single-cell RNAseq clusters for IP and oRG (counts > 0.02):

```python
EOMES_detected = np.where(sum(np.array([adata[adata.obs['AOI_type'] == 'EOMESpos',:].X[i,subset] > adata.obs['GeoLOD2.5_01'].loc[adata.obs['AOI_type'] == 'EOMESpos'].iloc[i]
for i in range(sum(adata.obs['AOI_type'] == 'EOMESpos'))]), axis = 0) > 1)[0]
HOPX_detected = np.where(sum(np.array([adata[adata.obs['AOI_type'] == 'HOPXpos',:].X[i,subset] > adata.obs['GeoLOD2.5_01'].loc[adata.obs['AOI_type'] == 'HOPXpos'].iloc[i]
for i in range(sum(adata.obs['AOI_type'] == 'HOPXpos'))]), axis = 0) > 1)[0]
print(len(EOMES_detected))
print(len(HOPX_detected))
IP_detected = np.where(polioudakis['IP'].loc[subset] > 0.02)[0]
oRG_detected = np.where(polioudakis['oRG'].loc[subset] > 0.02)[0]
print(len(IP_detected))
print(len(oRG_detected))
```

```python
allGenes = np.unique(np.concatenate([adata.var.index[subset][EOMES_detected], adata.var.index[subset][IP_detected], 
                     adata.var.index[subset][HOPX_detected], adata.var.index[subset][HOPX_detected]]))
```

```python
detectionTable = pd.DataFrame(data = False, columns = ['Detected in IP cluster', 'Detected in EOMESpos AOIs', 'Detected in oRG cluster', 'Detected in HOPXpos AOIs'],
            index = allGenes)
```

```python
detectionTable['Detected in IP cluster'] = [i in adata.var.index[subset][IP_detected] for i in detectionTable.index]
detectionTable['Detected in EOMESpos AOIs'] = [i in adata.var.index[subset][EOMES_detected] for i in detectionTable.index]
detectionTable['Detected in oRG cluster'] = [i in adata.var.index[subset][oRG_detected] for i in detectionTable.index]
detectionTable['Detected in HOPXpos AOIs'] = [i in adata.var.index[subset][HOPX_detected] for i in detectionTable.index]
```

```python
detectionTable.to_csv('/nfs/team283/aa16/KR_NAS/0_publication_tables/SupplementaryTableX_AllReceptorLigandsDetection.csv')
```

Plot detected in both, detected in single-cell RNAseq only, detected in wta only and not detected.

```python
overlapIP = np.intersect1d(EOMES_detected, IP_detected)
print(len(overlapIP))
uniqueEOMES = [EOMES_detected[i] for i in range(len(EOMES_detected)) if EOMES_detected[i] not in IP_detected]
print(len(uniqueEOMES))
uniqueIP = [IP_detected[i] for i in range(len(IP_detected)) if IP_detected[i] not in EOMES_detected]
print(len(uniqueIP))
undetectedIP = [range(sum(subset))[i] for i in range(sum(subset)) if range(sum(subset))[i] not in
             np.concatenate([IP_detected, EOMES_detected])]
print(len(undetectedIP))

overlapORG = np.intersect1d(HOPX_detected, oRG_detected)
print(len(overlapORG))
uniqueHOPX = [HOPX_detected[i] for i in range(len(HOPX_detected)) if HOPX_detected[i] not in oRG_detected]
print(len(uniqueHOPX))
uniqueORG = [oRG_detected[i] for i in range(len(oRG_detected)) if oRG_detected[i] not in HOPX_detected]
print(len(uniqueORG))
undetectedORG = [range(sum(subset))[i] for i in range(sum(subset)) if range(sum(subset))[i] not in
             np.concatenate([oRG_detected, HOPX_detected])]
print(len(undetectedORG))
```

Are WTA only genes really WTA only? Let's check if they are higher expressed in reference rings or expressed in some other scRNAseq cell types:

```python
roi_relations = pickle.load(open('data/ROI_relations.p', 'rb'))

uniqueHOPX1 = np.where(sum([adata.layers['CPM'][:,subset][:,uniqueHOPX][roi_relations['Ring'].iloc[i],:] <
adata.layers['CPM'][:,subset][:,uniqueHOPX][roi_relations['HOPXpos'].iloc[i],:]
for i in range(len(roi_relations['HOPXpos']))], axis = 0)/len(roi_relations['HOPXpos']) >= 0.5)[0]

uniqueEOMES1 = np.where(sum([adata.layers['CPM'][:,subset][:,uniqueEOMES][roi_relations['Ring'].iloc[i],:] <
adata.layers['CPM'][:,subset][:,uniqueEOMES][roi_relations['EOMESpos'].iloc[i],:]
for i in range(len(roi_relations['EOMESpos']))], axis = 0)/len(roi_relations['EOMESpos']) >= 0.5)[0]

uniqueHOPX_genes = adata.var_names[subset][uniqueHOPX][uniqueHOPX1]
uniqueEOMES_genes = adata.var_names[subset][uniqueEOMES][uniqueEOMES1]
uniqueHOPX_genes1 = adata.var_names[subset][uniqueHOPX]
uniqueEOMES_genes1 = adata.var_names[subset][uniqueEOMES]
uniqueHOPX_genes2 = uniqueHOPX_genes[sum(polioudakis.loc[uniqueHOPX_genes,:] > 0.02, axis = 1) == 0]
uniqueEOMES_genes2 = uniqueEOMES_genes[sum(polioudakis.loc[uniqueEOMES_genes,:] > 0.02, axis = 1) == 0]
uniqueHOPX_genes3 = uniqueHOPX_genes1[sum(polioudakis.loc[uniqueHOPX_genes1,:] > 0.02, axis = 1) == 0]
uniqueEOMES_genes3 = uniqueEOMES_genes1[sum(polioudakis.loc[uniqueEOMES_genes1,:] > 0.02, axis = 1) == 0]
```

```python
bars1 = [len(overlapIP), len(overlapORG)]
bars2 = [len(uniqueEOMES_genes), len(uniqueHOPX_genes)]
bars3 = [len(uniqueIP), len(uniqueORG)]
bars4 = [len(uniqueEOMES_genes3), len(uniqueHOPX_genes3)]
bars5 = [len(uniqueEOMES_genes2), len(uniqueHOPX_genes2)]
```

```python
detectionTable['Unique for EOMESpos AOIs  (filter by reference ring)'] = [i in uniqueEOMES_genes3 for i in detectionTable.index]
detectionTable['Unique for EOMESpos AOIs  (filter by reference ring and RNAseq)'] = [i in uniqueEOMES_genes2 for i in detectionTable.index]
detectionTable['Unique for HOPXpos AOIs  (filter by reference ring)'] = [i in uniqueHOPX_genes3 for i in detectionTable.index]
detectionTable['Unique for HOPXpos AOIs  (filter by reference ring and RNAseq)'] = [i in uniqueHOPX_genes2 for i in detectionTable.index]
```

Calculate average expression in HOPXpos and EOMESpos AOIs in 19pcw samples:

```python
eomes_average_cpm = np.mean(adata[adata.obs['AOI_type'] == 'EOMESpos',:].layers['CPM_corrected'], axis = 0)
hopx_average_cpm = np.mean(adata[adata.obs['AOI_type'] == 'HOPXpos',:].layers['CPM_corrected'], axis = 0)
```

Compare the two in a scatter plot:

```python
SMALL_SIZE = 18
MEDIUM_SIZE = 18
BIGGER_SIZE = 18

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=14)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=12)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

fig, ax = plt.subplots(1,1,figsize=(5,5))

pseudocount = 10**(-3)
ax.scatter(polioudakis['IP'].loc[subset].iloc[np.concatenate([IP_detected, EOMES_detected])]+pseudocount,
            eomes_average_cpm[subset][np.concatenate([IP_detected, EOMES_detected])], s = 2, c = 'blue',
            label = ('EOMESpos AOIs/ IP cluster'))
plt.scatter(polioudakis['vRG'].loc[subset].iloc[np.concatenate([oRG_detected, HOPX_detected])],
            hopx_average_cpm[subset][np.concatenate([oRG_detected, HOPX_detected])], s = 2, c = 'red',
            label = ('HOPXpos AOIs/ oRG cluster'))
ax.set_ylabel('Mean CPM in AOIs')
ax.set_xlabel('Mean counts in RNAseq cluster')
ax.set_xlim(pseudocount,3.5)
ax.set_xscale('log')
ax.set_yscale('log')
ax.legend()
rho1 = np.round(np.corrcoef(np.log(polioudakis['IP'].loc[subset].iloc[np.concatenate([IP_detected, EOMES_detected])]+1),
      np.log(eomes_average_cpm[subset][np.concatenate([IP_detected, EOMES_detected])]+1))[0,1],2)
rho2 = np.round(np.corrcoef(np.log(polioudakis['oRG'].loc[subset].iloc[np.concatenate([oRG_detected, HOPX_detected])]+1),
      np.log(hopx_average_cpm[subset][np.concatenate([oRG_detected, HOPX_detected])]+1))[0,1],2)
plt.text(0.29, 10, r'$\rho$ = ' + str(rho1), c = 'blue')
plt.text(0.29, 15, r'$\rho$ = ' + str(rho2), c = 'red')
plt.tight_layout()
plt.savefig('0_publication_figures/Figure5_LigandReceptors_Correlation_logScale.pdf')
```
