Map subtypes to 19pcw and 14pcw data. Save the results and produce cell type maps.

Load libraries and data:

```python
import sys,os
path = '/nfs/team283/aa16/KR_NAS/'
os.chdir(path)
sys.path.append(path)
```

```python
import pickle
import anndata
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
data_type = 'float32'
os.environ["THEANO_FLAGS"] = 'device=cuda,floatX=' + data_type + ',force_device=True' + ',dnn.enabled=False'
import isctools
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
```

```python
adata_wta = pickle.load(open(path + "data/nanostringWTA_fetailBrain_AnnData.p", "rb" ))
```

```python
adata_wta = adata_wta[adata_wta.obs['GoodQuality'],:]
```

```python
meanExpression_sc = pd.read_csv(path + "../KR_NAS/data/subclusterMeans.csv", index_col=0)
```

Make sure we have the same genes in reference and spatial data:

```python
wta_genes = np.asarray(adata_wta.var_names)
common_genes = [wta_genes[i] for i in range(len(wta_genes)) if wta_genes[i] in meanExpression_sc.index] 
adata_wta = adata_wta[:,[adata_wta.var_names[i] in common_genes + ['NegProbe-WTX'] for i in range(len(adata_wta.var_names))]]
meanExpression_sc = meanExpression_sc.reindex(index = common_genes)
```

Fit each slide seperately:

```python
slides = np.array(('00MU', ('00MV', '00MV-2'), '00MW'))
adata_slide = list()
for j in range(len(slides)):
    adata_slide.append(adata_wta[[adata_wta.obs['slide'][i] in slides[j] for i in range(len(adata_wta.obs['slide']))]])
    print(np.shape(adata_slide[j]))
```

```python
counts_geneProbes = list()
counts_negativeProbes = list()
counts_nuclei = list()

for i in range(len(slides)):
    counts_geneProbes.append(np.asarray(adata_slide[i][:,np.array(adata_slide[i].var != 'NegProbe-WTX').squeeze()].X))
    counts_negativeProbes.append(np.asarray(adata_slide[i][:,np.array(adata_slide[i].var =='NegProbe-WTX').squeeze()].X))
    counts_nuclei.append(np.asarray(adata_slide[i].obs['nuclei']).reshape(len(adata_slide[i].obs['nuclei']),1))

celltypes = meanExpression_sc.keys()
meanExpression_sc = np.asarray(meanExpression_sc)
```

```python
celltypes
```

```python
models = list()
for i in range(len(slides)):
    models.append(isctools.ProbeCounts_CoLocationModel_Poisson(
            cell_state_mat = meanExpression_sc,
            X_data = counts_geneProbes[i],
            Y_data = counts_negativeProbes[i],
            data_type='float32',
            n_iter=20000,
            learning_rate=0.001,
            fact_names=celltypes,
            gene_level_prior={'mean': 1/2, 'sd': 1/8, 'mean_var_ratio': 0.25},
            cell_number_prior={'cells_per_spot': counts_nuclei[i], 'factors_per_spot': 6, 'combs_per_spot': 50},
            cell_number_var_prior={'cells_mean_var_ratio': 0.1, 'factors_mean_var_ratio': 1, 'combs_mean_var_ratio': 1}))
```

```python
for i in range(len(slides)):
    print('Fitting for slide: ', slides[i])
    models[i].fit_advi_iterative(n_iter = 50000, learning_rate = 0.001, n=2, method='advi')
```

Plot history and stability:

```python
for i in range(len(slides)):
    models[i].plot_history()
    plt.show()
    models[i].plot_history(25000)
    plt.show()
    models[i].evaluate_stability()
    plt.show()
```

Evaluate posterior and plot it vs data:

```python
for i in range(len(slides)):
    print(slides[i])
    models[i].sample_posterior(node='all', n_samples=1000, save_samples=False);
    isctools.ProbeCounts_CoLocationModel_Poisson.compute_expected(models[i])
    models[i].plot_posterior_mu_vs_data()
    plt.show()
```

Calculate the spot factors:

```python
for i in range(len(slides)):
    models[i].sample2df(node_name='spot_factors')
```

Save models just in case:

```python
pickle.dump(models, file = open( "0_publication_figures_scripts/models_subtypeMapping.p", "wb" ))
```

Overview plot of the cell locations in the 19pcw and 14pcw samples:

```python
subset_plot = list()
for i in range(len(slides)):
    subset_plot.append([adata_slide[i].obs['AOI_type'][j] == 'Geometric' and
                        adata_slide[i].obs['Radial_position'][j] == 2 for j in range(len(adata_slide[i].obs['AOI_type']))])
```

```python
labels = ('19pcw Technical Replicate 1', '19pcw Technical Replicate 2', '14pcw')
```

```python
fig, ax = plt.subplots(3,1, figsize=(6,15))

SMALL_SIZE = 16
MEDIUM_SIZE = 16
BIGGER_SIZE = 16
LEGEND_SIZE = 14

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=LEGEND_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

for slide_index in range(3):

    subset_rois = subset_plot[slide_index]

    results = models[slide_index].spot_factors_df

    results_normed = np.array(results)
    results_normed = (results_normed.T/[sum(results_normed[i,]) for i in range(len(results_normed[:,1]))]).T

    results_normed = np.asarray(results)

    order = 6

    cellColours = np.array(('green', 'blue', 'purple', 'yellow', 'red', 'green', 'blue', 'purple', 'yellow', 'red', 'green', 'blue', 'purple', 'yellow', 'red', 'green', 'blue', 'purple', 'yellow', 'red'))
    celltypes_to_plot = celltypes[50:55]
    for i in range(len(celltypes_to_plot)):
        x = adata_slide[slide_index].obs['VCDepth'][subset_rois]
        y = results_normed[:,celltypes == celltypes_to_plot[i]][:,0][subset_rois]
        ax[slide_index].plot(np.unique(x), np.poly1d(np.polyfit(x, y, order))(np.unique(x)), label = celltypes_to_plot[i],
                 c = cellColours[i])
        ax[slide_index].scatter(x, y, c = cellColours[i])
        ax[slide_index].set_xlabel('CorticalDepth')
        ax[slide_index].set_ylabel('Total Number')
        ax[slide_index].legend()
        ax[slide_index].set_title(labels[slide_index])
plt.tight_layout()
plt.savefig('0_publication_figures/Figure3_OPCSubtypeMapping_TotalNumbers_ScatterPlots.pdf')
plt.show()
```
