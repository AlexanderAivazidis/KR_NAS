```python
import sys, ast, os
import time
import pickle
import scanpy as sc
import anndata
import pandas as pd
import numpy as np
import os
from plotnine import *
import matplotlib.pyplot as plt 
import matplotlib
data_type = 'float32'
os.environ["THEANO_FLAGS"] = 'device=cuda,floatX=' + data_type + ',force_device=True' + ',dnn.enabled=False'
# /nfs/team283/vk7/software/miniconda3farm5/envs/cellpymc/bin/pip install git+https://github.com/vitkl/cell2location.git
path = '/nfs/team283/aa16/'
from matplotlib import rcParams
import seaborn as sns
# scanpy prints a lot of warnings
import warnings
warnings.filterwarnings('ignore')
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
```

Loading data and metadata

```python
# Let's load the data and metadata from our Nanostring experiment:

counts = pd.read_table(path + 'KR_NAS/data/Sanger_288ROIs_TargetCountMatrix.txt')
genes = counts['TargetName']
counts = counts.drop('TargetName',1)
counts = counts.rename(index=genes)
counts = counts.astype('int')
subset_genes = np.array([sum(counts.iloc[i,:]) for i in range(len(counts.iloc[:,0]))]) > 3 * 288
counts = counts[subset_genes]
metadata = pd.read_csv(path + 'KR_NAS/data/NanoString sequencing all annotations 2020.02.10.csv')
metadata = metadata.iloc[0:286,]
metadata = metadata.rename(index=metadata['Sanger_sampleID'])
metadata = metadata.reindex(np.array(counts.columns))
properties = pd.read_table(path + 'KR_NAS/data/Sanger_288ROIs_SegmentProperties.txt')
properties = properties.rename(index=properties['DSP_Sample_ID'])
properties = properties.reindex(np.array(metadata['Sample_ID']))
columnNames = ('x', 'y', 'total_counts', 'Q3_counts')
sample_info = pd.DataFrame(index=metadata['Sample_ID'], columns=columnNames)
sample_info['x'] = np.array(metadata['VCDepth'])
sample_info['y'] = np.array(metadata['Radial_position'])
sample_info['total_counts'] = [sum(counts.iloc[:,i]) for i in range(len(counts.iloc[1,:]))] 
sample_info['Q3_counts'] = [sum(np.sort(counts.iloc[:,i])[int(np.round(0.5*len(counts.iloc[:,i]))):int(np.round(0.75*len(counts.iloc[:,i])))]) for i in range(len(counts.iloc[1,:]))]
```

Choose subsets of ROIs:

```python
# Choose which ROIs and genes to use, e.g. top 100 markers of each cell type:
rSlides1 = np.array(('00MU', '00MV', '00MV-2'))
rSlides2 = np.array(('00MU', '00MV', '00MV-2', '00MR'))

subset1 = np.where([metadata['slide'][i] in rSlides1.flatten() 
                   and metadata['AOI_type'][i] in ('Geometric')
                            for i in range(len(metadata['slide']))])[0]

subset2 = np.where([metadata['slide'][i] in rSlides2.flatten() 
                   and metadata['AOI_type'][i] in ('Geometric')
                            for i in range(len(metadata['slide']))])[0]
```

Calculate correlation in expression of normalized counts:

```python
normCounts = np.array([counts.iloc[:,i]/sum(counts.iloc[:,i]) for i in range(len(counts.iloc[1,:]))]).T*10**(5)
counts_subset1 = normCounts[:,subset1]
counts_subset2 = normCounts[:,subset2]
metadata_subset1 = metadata.iloc[subset1,:]
metadata_subset2 = metadata.iloc[subset2,:]
correlationMatrix1 = np.corrcoef(np.log2(np.array(counts_subset1)).T)
correlationMatrix2 = np.corrcoef(np.log2(np.array(counts_subset2)).T)
plt.imshow(correlationMatrix1)
plt.show()
plt.imshow(correlationMatrix2)
```

Set diagonal to zero:

```python
np.fill_diagonal(correlationMatrix1,0)
np.fill_diagonal(correlationMatrix2,0)
plt.imshow(correlationMatrix1)
plt.show()
plt.imshow(correlationMatrix2)
plt.show()
```

Subset correlation matrix to slides we want to compare:

```python
correlationMatrix1_subset = correlationMatrix1[metadata_subset1['slide'] == '00MU', :]
correlationMatrix1_subset = correlationMatrix1_subset[:,[metadata_subset1['slide'][i] in ('00MV', '00MV-2') for i in range(len(metadata_subset1['slide']))]]
roi_names1 = np.array((np.array(metadata_subset1.index)[metadata_subset1['slide'] == '00MU'],
                      np.array(metadata_subset1.index)[[metadata_subset1['slide'][i] in ('00MV', '00MV-2') for i in range(len(metadata_subset1['slide']))]])) 
plt.imshow(correlationMatrix1_subset)
plt.show()
correlationMatrix2_subset = correlationMatrix2[metadata_subset2['slide'] == '00MR', :]
correlationMatrix2_subset = correlationMatrix2_subset[:,[metadata_subset2['slide'][i] in ('00MV', '00MV-2', '00MU') for i in range(len(metadata_subset2['slide']))]]
roi_names2 = np.array((np.array(metadata_subset2.index)[metadata_subset2['slide'] == '00MR'],
                      np.array(metadata_subset2.index)[[metadata_subset2['slide'][i] in ('00MV', '00MV-2', '00MU') for i in range(len(metadata_subset2['slide']))]])) 

plt.imshow(correlationMatrix2_subset)
plt.show()
```

Extract most correlated ROIs between slides:

```python
cor1 = np.round(np.amax(correlationMatrix1_subset),2)
print(np.amax(correlationMatrix1_subset))
index1 = np.unravel_index(correlationMatrix1_subset.argmax(), correlationMatrix1_subset.shape)
chosen_rois1 = np.array((roi_names1[0][index1[0]], roi_names1[1][index1[1]]))
print(chosen_rois1[0])
print(chosen_rois1[1])
print(metadata['slide'].loc[chosen_rois1[0]])
print(metadata['slide'].loc[chosen_rois1[1]])
print(metadata['VCDepth'].loc[chosen_rois1[0]])
print(metadata['VCDepth'].loc[chosen_rois1[1]])

cor2 = np.round(np.amax(correlationMatrix2_subset),2)
print(np.amax(correlationMatrix2_subset))
index2 = np.unravel_index(correlationMatrix2_subset.argmax(), correlationMatrix2_subset.shape)
chosen_rois2 = np.array((roi_names2[0][index1[0]], roi_names2[1][index2[1]]))
print(chosen_rois2[0])
print(chosen_rois2[1])
print(metadata['slide'].loc[chosen_rois2[0]])
print(metadata['slide'].loc[chosen_rois2[1]])
print(metadata['VCDepth'].loc[chosen_rois2[0]])
print(metadata['VCDepth'].loc[chosen_rois2[1]])
```

Finally plot normalized counts to show correlation:

```python
alpha_value = 0.1
fig, ax = plt.subplots(1,2,figsize=(10, 5))
ax[0].title.set_text('19pcw Cortical Plate Gene Expression: \n Technical Replicates')
ax[0].scatter(np.log2(counts.loc[:,chosen_rois1[0]]/sum(counts.loc[:,chosen_rois1[0]])*10**5),
              np.log2(counts.loc[:,chosen_rois1[1]]/sum(counts.loc[:,chosen_rois1[1]])*10**5),
             10,alpha = alpha_value)
ax[0].set_xlabel('Log2 Normalized Gene Counts')
ax[0].set_ylabel('Log2 Normalized Gene Counts')
ax[0].text(0.25, 0.75,r'$\rho$' + '=' + str(cor1),
     horizontalalignment='center',
     verticalalignment='center',
     transform = ax[0].transAxes)
ax[1].title.set_text('19 pcw Cortical Plate Gene Expression: \n Biological Replicates')
ax[1].scatter(np.log2(counts.loc[:,chosen_rois2[0]]/sum(counts.loc[:,chosen_rois2[0]])*10**5),
              np.log2(counts.loc[:,chosen_rois2[1]]/sum(counts.loc[:,chosen_rois2[1]])*10**5),
             10, alpha = alpha_value)
ax[1].set_xlabel('Log2 Normalized Gene Counts')
ax[1].set_ylabel('Log2 Normalized Gene Counts')
ax[1].text(0.25, 0.75,r'$\rho$' + '=' + str(cor2),
     horizontalalignment='center',
     verticalalignment='center',
     transform = ax[1].transAxes)
fig.savefig('GeoMxReproducibility.pdf')
```

Also plot reproducibility for cell type specific AOIs:

```python
adata = pickle.load(open('../data/nanostringWTA_fetailBrain_AnnData.p',  "rb"))
adata = adata[:,adata.var_names != 'NegProbe-WTX']
```

Choose AOIs:

```python
fig, ax = plt.subplots(2,2, figsize = (10,10))
for i, slide in enumerate((('00MU'), ('00MV', '00MV-2'))):
    for j, aoi_type in enumerate(('EOMESpos', 'HOPXpos')):
        subset = [adata.obs['slide'].iloc[i] in slide and
                                 adata.obs['AOI_type'].iloc[i] == aoi_type
                             for i in range(len(adata.obs['slide']))]
        colours = np.repeat('black', sum(subset))
        chosen_rois = [adata.obs['slide'].iloc[i] in slide
                       and adata.obs['AOI_type'].iloc[i] == aoi_type
                       and np.abs((adata.obs['VCDepth'].iloc[i] - 0.3)) < 0.05
                             for i in np.where(subset)[0]]
        colours[chosen_rois] = 'red'
        ax[i,j].set_title(aoi_type + ', slide: ' + ''.join(slide))
        ax[i,j].scatter(adata.obs['VCDepth'].loc[subset], adata.obs['Radial_position'].loc[subset],
                    s = 1000,
                    marker = 's',
                    c = colours)
        ax[i,j].set_ylabel('Radial_position')
        ax[i,j].set_xlabel('Cortical Depth')
        ax[i,j].set_ylim(0,5)
        ax[i,j].set_xlim(0,0.45)

plt.tight_layout()
fig.savefig('../../KR_NAS/0_publication_figures/Supplementary_AOI-WTAReproducibility_ChosenAOIs.pdf')
plt.show()
```

Calculate and plot correlation:

```python
aoi_positions = pd.DataFrame(columns = ('index', 'slide', 'radial_position', 'VCDepth', 'aoi_type'), index = range(8))
```

```python
count = 0
for slide in (('00MU'), ('00MV', '00MV-2')):
    for aoi_type in ('EOMESpos', 'HOPXpos'):
        for radial_position in (1,3):
            aoi_positions.iloc[count,0] = np.where([adata.obs['slide'].iloc[i] in slide
                                   and adata.obs['AOI_type'].iloc[i] == aoi_type
                                   and adata.obs['Radial_position'].iloc[i] == radial_position
                                   and np.abs((adata.obs['VCDepth'].iloc[i] - 0.3)) < 0.05
                                   for i in range(len(adata.obs['slide']))])[0][0]
            aoi_positions.iloc[count,1] = adata.obs['slide'].iloc[aoi_positions.iloc[count,0]]
            aoi_positions.iloc[count,2] = adata.obs['Radial_position'].iloc[aoi_positions.iloc[count,0]]
            aoi_positions.iloc[count,3] = adata.obs['VCDepth'].iloc[aoi_positions.iloc[count,0]]
            aoi_positions.iloc[count,4] = adata.obs['AOI_type'].iloc[aoi_positions.iloc[count,0]]
            count += 1
aoi_positions
```

```python
fig, ax = plt.subplots(4,4, figsize = (10, 10))
correlations = pd.DataFrame(columns = range(4), index = range(4))
for i, ii in enumerate((0,1,4,5)):
    for j, jj in enumerate((0,1,4,5)):
        correlations.iloc[i,j] = np.round(np.corrcoef(np.log2(adata.layers['CPM'][aoi_positions['index'].iloc[ii],:]),
                    np.log2(adata.layers['CPM'][aoi_positions['index'].iloc[jj],:])),2)[0,1]
        
        ax[i,j].set_title('log2(CPM), ' +  r'$\rho$' + '= ' + str(correlations.iloc[i,j]), size = 10)
        ax[i,j].scatter(np.log2(adata.layers['CPM'][aoi_positions['index'].iloc[ii],:]),
              np.log2(adata.layers['CPM'][aoi_positions['index'].iloc[jj],:]),
             10,alpha = alpha_value)
        ax[i,j].set_xlabel(aoi_positions['aoi_type'].iloc[ii] + ', ' +
                           aoi_positions['slide'].iloc[ii] + ', ' + str(int(aoi_positions['radial_position'].iloc[ii])),
                          size = 10)
        ax[i,j].set_ylabel(aoi_positions['aoi_type'].iloc[jj] + ', ' +
                           aoi_positions['slide'].iloc[jj] + ', ' + str(int(aoi_positions['radial_position'].iloc[jj])),
                          size = 10)

plt.tight_layout()
fig.savefig('../../KR_NAS/0_publication_figures/Supplementary_AOI-WTAReproducibility_EOMESpos.png')
```
